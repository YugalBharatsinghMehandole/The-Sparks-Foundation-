# Step 5 :- Categorical Analysis of the Data

## Ship Mode

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
sns.countplot(x='Ship Mode',data=data,palette='cubehelix')
plt.title('Ship Mode count plot')
sns.despine()
shipmodecounts=data['Ship Mode'].value_counts()
plt.subplot(1,2,2)
plt.pie(shipmodecounts,labels=shipmodecounts.index,shadow=True,colors=['green','pink','lightblue','purple'],explode=[0.1,0,0,0.2],autopct='%1.1f%%')
plt.legend()
```

#### Highest number of customers are opting for Standard class ship mode
#### Least number of customers are opting for same day ship mode.

## Customer Segment

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
sns.countplot(x='Segment',data=data,palette='Spectral')
plt.title('Customer segment count plot')
sns.despine()

segmentcounts=data['Segment'].value_counts()
plt.subplot(1,2,2)
plt.pie(segmentcounts,labels=segmentcounts.index,colors=['purple','lime','orange'],shadow=True,explode=[0.1,0,0],autopct='%1.1f%%')
plt.legend()
```

### We have highest number of buyers in the Consumer segment as compared to Corporate or Home office. So we can say that Consumers are our most potential buyers.

## Region

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
sns.countplot(x='Region',data=data,palette='icefire')
plt.title('Region Sale count plot')
sns.despine()

regioncount=data['Region'].value_counts()
plt.subplot(1,2,2)
plt.pie(regioncount,labels=regioncount.index,shadow=True,colors=['darksalmon','lightblue','yellow','pink'],explode=[0.1,0.1,0,0],autopct='%1.1f%%')
plt.legend()
```

### The above plots tells us that West and East region contributes to our sales. So, we need to work on South region to increse the revenue.

## Product Category

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
sns.countplot(x='Category',data=data,palette='rocket')
plt.title('Category Sale count plot')
sns.despine()

categorycounts=data['Category'].value_counts()
plt.subplot(1,2,2)
plt.pie(categorycounts,labels=categorycounts.index,shadow=True,colors=['lightblue','yellow','green'],explode=[0.1,0,0],autopct='%1.1f%%')
plt.legend()
```

### We can see that Office supplies are the most bought items from above plot. Now we can look in other sectors like Furniture and technology to increase Revenue.
